```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df = pd.read_csv('mental_health_finaldata_1.csv')
df.head()
```

```python

encoder = LabelEncoder()
category = df.select_dtypes(include=['object','category'])
for x in category:
    df[x] = encoder.fit_transform(df[x])
df.head()
```

```python
genders = ['Male','Female']
Growing_Stress = ['Yes', 'No']
```

```python
data = {'Age': ['20-25', '30-above', '30-above', '25-30', '16-20'],
        'Days_Indoors': ['1-14days', '31-60days', 'Go out Everyday', '1-14days', 'more than 2 months']
}
df = pd.DataFrame(data)
sns.boxplot(x='Age', y='Days_Indoors', data=df, order =['20-25', '30-above', '25-30', '16-20'])
plt.xlabel('Age')
plt.ylabel('Days_Indoors')
plt.title('Box Plot: Age vs Days Indoors')

plt.show()
```

```python
data = {'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
        'Growing_Stress': ['Yes', 'Yes', 'No', 'Yes', 'Yes']}

df = pd.DataFrame(data)

sns.countplot(x='Growing_Stress', hue='Gender', data=df, order=['Yes', 'No'], palette='muted')
plt.xlabel('Growing_Stress')
plt.ylabel('Count')
plt.title('Gender-Based Differences in Growing Stress')

plt.show()
```
